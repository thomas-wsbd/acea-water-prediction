```python
import utils
import geopandas as gpd

if os.path.exists('./data/geolocations.geojson'):
    gdf = gpd.read_file('./data/geolocations.geojson')
else:
    gdf = utils.returngeo()
```

```python
# for the map
gdf['lat'] = gdf['geometry'].y
gdf['lon'] = gdf['geometry'].x

import folium
icons = {'Rainfall': 'tint', 'Temperature': 'sun-o', 'target': 'flag', 'Both': 'bar-chart'} # rainfall cloud, temperature sun, both (temperature & rainfall) flag
colors = {'Auser': 'cadetblue', 'Luco': 'lightblue', 'Doganella': 'blue', 'Petrignano': 'darkblue', 'Bilancino': 'darkgreen', 'Arno': 'orange', 'Amiata': 'pink', 'Madonna di Canneto': 'purple', 'Lupa': 'darkpurple'} # blues aquifers, green lake, orange river, pinks water springs

m = folium.Map(location=[43.15, 12.07], zoom_start=7, tiles='cartodbpositron')
for i, row in gdf.iterrows():
    folium.Marker([row['lat'], row['lon']], tooltip='<b>name:</b> {}<br><b>parameters:</b> {}<br><b>adress:</b> {}<br><b>waterbody:</b> {}'.format(row['place'], row['param'], row['adress'], row['feat']), icon=folium.Icon(color=colors.get(row['feat']), icon=icons.get(row['param']), prefix='fa')).add_to(m)

m
```
